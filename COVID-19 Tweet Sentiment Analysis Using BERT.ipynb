#<b>2. Process</b>

##<b>2.1 Case Folding</b>

```python
df_casefold = df_drop_subset.copy()

# Case Folding (Lowercase)
df_casefold['text_lower'] = df_casefold['text'].str.lower()
df_casefold.head()
```

##<b>2.2 Tweet Cleaning & Tokenizing</b>

```python
import pandas as pd
import re
from nltk.tokenize import RegexpTokenizer

def clean_tweet_text(df, text_column):
    df_cleaned = df.copy()

    # Menghapus URL
    df_cleaned['text_no_url'] = df_cleaned[text_column].apply(lambda x: re.sub(r'http\S+|www\S+|https\S+', '', x, flags=re.MULTILINE))

    # Menghapus mentions (@username)
    df_cleaned['text_no_mentions'] = df_cleaned['text_no_url'].apply(lambda x: re.sub(r'@\w+', '', x))

    # Menghapus hashtag
    df_cleaned['text_no_hashtags'] = df_cleaned['text_no_mentions'].apply(lambda x: re.sub(r'#\w+', '', x))

    # Menghapus angka
    df_cleaned['text_no_numbers'] = df_cleaned['text_no_hashtags'].apply(lambda x: re.sub(r'\d+', '', x))

    # Menghapus tanda baca dan tokenisasi
    tokenizer = RegexpTokenizer(r'\w+')  # Hanya mengambil alphanumeric characters
    df_cleaned['text_token'] = df_cleaned['text_no_numbers'].apply(tokenizer.tokenize)

    return df_cleaned['text_token']


df_token = df_casefold.copy()
df_token['text_token'] = clean_tweet_text(df_casefold, 'text_lower')
df_token.head()
```

##<b>2.3 Stopwords Removal</b>

```python
import nltk

# Download daftar stopword di nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
from nltk.corpus import stopwords

# Menampilkan daftar stopword
print(f"Daftar Stopword Pada Library NLTK:\n{stopwords.words('english')}\n")
print(f"Jumlah Stopword Pada Library NLTK:\n{len(stopwords.words('english'))}")
```

```python
stop_words = set(stopwords.words('english'))

# Stopwords Removal
def sw_remove(words):
  filtered_sentence = []
  for w in words:
    if w not in stop_words:
      filtered_sentence.append(w)
  return filtered_sentence

df_sword = df_token.copy()
df_sword['text_swremove'] = df_sword['text_token'].apply(sw_remove)
df_sword.head()
```

##<b>2.4 Lemmatization</b>

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

# Lemmatization
def lemmatize_text(tokens):
    lem_tokens = []
    for token in tokens:
        lem_token = lemmatizer.lemmatize(token)
        lem_tokens.append(lem_token)
    return lem_tokens

df_lemma = df_sword.copy()
df_lemma['text_lemma'] = df_lemma['text_swremove'].apply(lemmatize_text)
df_lemma.head()
```

```python
df_covid = df_lemma.copy()
df_covid['text_clean'] = df_lemma['text_lemma'].apply(lambda x: ' '.join(x))
df_covid
```

##<b>2.5 Load Model</b>

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

```python
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

##<b>2.6 Get Sentiment Score</b>

```python
def sentiment_score(review):
    tokens = tokenizer.encode(review, return_tensors='pt')
    result = model(tokens)
    return int(torch.argmax(result.logits))+1
```

```python
df_final = df_covid[['user_name',	'user_location', 'user_followers',	'user_friends',	'user_favourites',
                     'user_verified',	'date',	'text',	'hashtags', 'is_retweet', 'text_clean']].copy()
```

```python
df_final
```

```python
df_final['sentiment_score'] = df_final['text_clean'].apply(lambda x: sentiment_score(x[:512]))
df_final.head()
```

##<b>3.1 Sentiment Label</b>

```python
# Fungsi untuk menetapkan label berdasarkan skor
def label_score(score):
    if score in [4, 5]:
        return 'positive'
    elif score == 3:
        return 'neutral'
    elif score in [1, 2]:
        return 'negative'
    else:
        return 'unknown'

# Menambahkan kolom label ke DataFrame
df_final['sentiment_label'] = df_final['sentiment_score'].apply(label_score)
df_final.head()
```

```python
from collections import Counter

# Menghitung frekuensi kata untuk masing-masing sentimen
sentiment_word_counts = {}

for sentiment in df_final['sentiment_label'].unique():
    sentiment_text = ' '.join(df_final[df_final['sentiment_label'] == sentiment]['text_clean'])
    tokens = sentiment_text.split()
    word_counts = Counter(tokens)
    sentiment_word_counts[sentiment] = word_counts

# Membuat grafik untuk masing-masing sentimen dengan warna berbeda
plt.figure(figsize=(15, 10))

colors = {'negative': 'red', 'positive': 'green', 'neutral': 'yellow'}

for i, (sentiment, word_counts) in enumerate(sentiment_word_counts.items(), 1):
    plt.subplot(1, 3, i)
    top_words = dict(word_counts.most_common(20))
    plt.barh(list(top_words.keys()), list(top_words.values()), color=colors[sentiment])
    plt.gca().invert_yaxis()  # Memflip sumbu y untuk menampilkan kata yang paling sering muncul di bagian atas
    plt.title(f"Top 20 Kata - {sentiment.capitalize()} Sentiment")
    plt.xlabel('Jumlah Kemunculan')
    plt.ylabel('Kata')

plt.tight_layout()
plt.show()
```

```python
from collections import Counter

# Menghitung frekuensi kata untuk masing-masing sentimen
sentiment_word_counts = {}

for sentiment in df_final['sentiment_label'].unique():
    sentiment_text = ' '.join(df_final[df_final['sentiment_label'] == sentiment]['text_clean'])
    tokens = sentiment_text.split()
    word_counts = Counter(tokens)
    sentiment_word_counts[sentiment] = word_counts

# Membuat kamus untuk masing-masing sentimen
sentiment_word_dict = {}

for sentiment, word_counts in sentiment_word_counts.items():
    top_words = {word: count for word, count in word_counts.items() if all(word not in sentiment_word_counts[other_sentiment] for other_sentiment in sentiment_word_counts if other_sentiment != sentiment)}
    # Mengurutkan kamus berdasarkan jumlah kemunculan (dalam urutan menurun)
    sorted_words = dict(sorted(top_words.items(), key=lambda item: item[1], reverse=True)[:20])
    sentiment_word_dict[sentiment] = sorted_words


print("Top 20 Kata Masing-Masing Sentimen:\n")
# Menampilkan hasil dalam format yang diinginkan
for sentiment, word_dict in sentiment_word_dict.items():
    print(f"{sentiment.capitalize()} Sentiment:")
    for word, count in word_dict.items():
        print(f"'{word}': {count}")
    print()
```

```python
# Fungsi untuk membagi user_location ke dalam kategori yang diinginkan
def categorize_location(location):
    if 'India' in location or 'New Delhi' in location or '%islike%' in location:
        return 'India'
    elif 'UK' in location or 'United Kingdom' in location or 'London' in location or '%islike%' in location:
        return 'UK'
    elif 'US' in location or 'United States' in location or '%islike%' in location:
        return 'US'
    elif 'Canada' in location or '%islike%' in location:
        return 'Canada'
    elif 'Switzerland' in location or '%islike%' in location:
        return 'Switzerland'
    else:
        return 'Another_country'

df_country = df_final.copy()

# Terapkan fungsi untuk membuat kolom baru 'location_category'
df_country['location_category'] = df_country['user_location'].apply(categorize_location)
df_country.head()
```

```python
import seaborn as sns

# Menghitung jumlah tweet untuk setiap kombinasi negara dan jenis sentiment
sentiment_counts = df_country.groupby(['location_category', 'sentiment_label']).size().unstack(fill_value=0)

# Mengubah data ke dalam format yang cocok untuk plotting
sentiment_counts = sentiment_counts.reset_index().melt(id_vars='location_category', var_name='sentiment_label', value_name='count')

# Membuat grafik
plt.figure(figsize=(12, 8))
barplot = sns.barplot(data=sentiment_counts, x='location_category', y='count', hue='sentiment_label', palette={'negative': 'red', 'neutral': 'yellow', 'positive': 'green'})

# Menambahkan label jumlah di atas setiap batang
for p in barplot.patches:
    height = p.get_height()
    barplot.annotate(format(height, '.1f'),
                     (p.get_x() + p.get_width() / 2., height),
                     ha = 'center', va = 'center',
                     xytext = (0, 9),
                     textcoords = 'offset points')

# Menambahkan judul dan label
plt.title('Jumlah Tweet Berdasarkan Sentimen di Tiap Negara')
plt.xlabel('Negara')
plt.ylabel('Jumlah Tweet')
plt.legend(title='Sentimen')
plt.show()
```

Dari grafik tersebut, dapat ditarik beberapa interpretasi terkait dengan sentimen yang diekspresikan terhadap beberapa negara:

Canada: Lebih banyak tweet dengan sentimen negatif (78) daripada positif (64) dan netral (22), menunjukkan ada kecenderungan untuk mengekspresikan ketidakpuasan atau keprihatinan di antara pengguna terhadap topik terkait Kanada.

India: Terdapat jumlah tweet yang hampir seimbang antara positif (645) dan negatif (639), dengan jumlah netral (102) yang juga signifikan. Hal ini menunjukkan adanya perbincangan yang luas dan bervariasi mengenai India, mencakup baik aspek positif maupun negatif.

Swiss: Negara lain dengan lebih banyak tweet positif (53) daripada negatif (35) dan netral (17), menunjukkan bahwa umumnya pendapat positif mendominasi pembicaraan tentang Swiss.

UK (Inggris): Terdapat lebih banyak tweet positif (312) dibandingkan dengan negatif (236) dan netral (54). Ini menunjukkan adanya sentimen positif yang lebih dominan terhadap topik yang berkaitan dengan Inggris.
